```python
import os
import re

labels_file = 'faceDS.txt'
dataraw_dir = 'rawdata'
output_file = 'cleared_faceDS.txt'

# 正则表达式用于匹配编号、性别和年龄（针对有效数据）
pattern_valid = re.compile(r'^(\d+)\s*\(_sex\s+\s*([^\)]+)\)\s*\(_age\s+\s*([^\)]+)\)')
# 正则表达式用于匹配缺失数据的编号
pattern_missing = re.compile(r'^(\d+)\s*\(_missing\s+descriptor\)')

# 存储有效数据的列表
valid_entries = []
# 存储缺失数据编号的集合
missing_image_ids = set()

# 读取标签文件并处理每一行
with open(labels_file, 'r') as f:
    for line in f:
        # 去除行尾的换行符和可能的前后空格
        line = line.strip()
        
        # 使用正则表达式匹配缺失数据的编号
        match_missing = pattern_missing.match(line)
        if match_missing:
            missing_id = match_missing.group(1)
            missing_image_ids.add(missing_id)
            continue  # 跳过缺失数据的进一步处理
        
        # 使用正则表达式匹配有效数据的编号、性别和年龄
        match_valid = pattern_valid.match(line)
        if match_valid:
            image_id = match_valid.group(1)
            sex = match_valid.group(2)
            age = match_valid.group(3)
            
            # 存储有效数据
            valid_entries.append(f"{image_id} ({sex} {age})\n")
        #else:
            # 如果不匹配有效数据格式，可能是其他格式错误的行，记录日志或处理
            #print(f"Warning: Line does not match expected valid or missing format: {line}")

# 写入新的标签文件
with open(output_file, 'w') as f:
    f.writelines(valid_entries)

# 删除dataraw中对应缺失数据的图片
for missing_id in missing_image_ids:
    # 构造图片路径（假设图片是jpg格式，且编号与文件名相同）
    image_path = os.path.join(dataraw_dir, f"{missing_id}")
    
    # 检查图片是否存在并删除
    if os.path.exists(image_path):
        os.remove(image_path)
        print(f"Deleted image: {image_path}")
    else:
        print(f"Warning: Image for ID {missing_id} not found.")
print("Processing complete. Filtered labels saved to:", output_file)
```
